```python
import pandas as pd
df = pd.read_csv('vi_dictionary.csv')
df
```

```python
# Filter for words that are nouns, 5-9 chars long, 2 words, no hyphens
filtered_df = df[
    # Filter for part_of_speech = "danh từ"
    (df['part_of_speech'] == 'danh từ') &
    # Filter for word length > 4 and < 10
    (df['word'].str.len() > 4) &
    (df['word'].str.len() < 10) &
    # Filter for words that split into exactly 2 parts
    (df['word'].str.split().str.len() == 2) &
    # Filter out words containing "-", handling NaN values
    (~df['word'].str.contains('-', na=False)) &
    # Ensure each individual word has length > 2
    (df['word'].apply(lambda x: all(len(word) > 2 for word in str(x).split())))
].reset_index(drop=True)

# Display the filtered dataframe
filtered_df.to_csv('filtered_words.csv', index=False)
filtered_df
```

```python
# Export words from filtered DataFrame to a text file
with open('filtered_words.txt', 'w', encoding='utf-8') as f:
    for word in filtered_df['word']:
        f.write(f"{word}\n")

print(f"Exported {len(filtered_df)} words to filtered_words.txt")
```

```python
import dotenv
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()
# Get the OpenAI API key from environment variables
api_key = os.getenv("GEMINI_API_KEY")
```

```python
import os
import pandas as pd
import numpy as np
import time

from google import genai
from google.genai import types

client = genai.Client(api_key=api_key)

# File to save embeddings
embeddings_file = 'word_embeddings.csv'

# Check for existing embeddings - resume mechanism
already_processed = set()
if os.path.exists(embeddings_file) and os.path.getsize(embeddings_file) > 0:
  try:
    existing_embeddings = pd.read_csv(embeddings_file)
    print(f"Found existing embeddings file: {embeddings_file}")
    print(existing_embeddings.head())
    # filtered_df[filtered_df.columns[0]].tolist()
    word_column = existing_embeddings[existing_embeddings.columns[0]].tolist()
    print(f"Found {len(already_processed)} already processed words")
  except pd.errors.EmptyDataError:
    print("Embeddings file exists but is empty")

# Get words to process (skip already processed ones)
words_to_process = [word for word in filtered_df['word'] if word not in already_processed]
print(f"Processing {len(words_to_process)} new words")

# Keep track of progress
total_words = len(words_to_process)
file_exists = os.path.exists(embeddings_file)

for i, word in enumerate(words_to_process):
  try:
    # Get embedding using the Gemini model
    result = client.models.embed_content(
      model="gemini-embedding-exp-03-07",
      contents=word,
      config=types.EmbedContentConfig(task_type="SEMANTIC_SIMILARITY",
                      output_dimensionality=1024),
    )
    print(result)
    # Extract embedding values
    embedding = result.embeddings[0].values
    # Create row with word and all embedding dimensions
    row_data = {'word': word}
    for dim, value in enumerate(embedding):
      row_data[f'dim_{dim}'] = value
    # Create a single-row DataFrame for this embedding
    word_df = pd.DataFrame([row_data])
    # Save directly to CSV - either create new file or append
    if not file_exists:
      word_df.to_csv(embeddings_file, index=False)
      file_exists = True
    else:
      word_df.to_csv(embeddings_file, mode='a', header=False, index=False)
    print(f"Embedded and saved: {word} [{i+1}/{total_words}]")
  except Exception as e:
    print(f"Error embedding '{word}': {str(e)}")
  # Small delay to avoid rate limits
  time.sleep(1)

print("Embedding process complete!")
```
